```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from IPython.display import Audio
```

A simple harmonic signal

```python
fs = 5000
T = 4
f1 = 440
f2 = 490
N = fs * T
x = np.sin(2 * np.pi * f1 * np.arange(N) / fs) + np.sin(2 * np.pi * f2 * np.arange(N) / fs)
Audio(data=x, rate=fs)
```

Periodogram $Sp_{xx}^N$ of $x$, using all the available samples

```python
freqs, SpxxN = signal.periodogram(x,fs=fs,return_onesided=False)
plt.figure()
plt.plot(freqs,SpxxN)
```

Periodogram $Sp_{xx}^{110}$ of $x$, using only $110$ samples

```python
freqs, SpxxN = signal.periodogram(x[1:110],fs=fs,return_onesided=False)
plt.plot(freqs,SpxxN)
```

Periodogram $Sp_{xx}^{90}$ of $x$, using only $90$ samples

```python
freqs, SpxxN = signal.periodogram(x[1:90],fs=fs,return_onesided=False)
plt.plot(freqs,SpxxN)
```
