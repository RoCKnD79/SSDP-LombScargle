# Numerical Exercise: AR Model

```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

#### AR Model

```python
ar = np.array([-.0009, -.0012, -.0011, -.0004, -.0001]) #np.array([1, -.75, .25, .5, .25])
ma = np.array([1])
ar_process = sm.tsa.ArmaProcess(ar, ma)
```

#### AR Sample Generation 

```python
N = 1000
ar_data = ar_process.generate_sample(nsample=N)
fig, ax = plt.subplots()
ax.set_title("AR process")
plt.plot(ar_data)
```

#### White Noise Sample Generation

```python
wn = np.random.normal(0,1,N)
fig, ax = plt.subplots()
ax.set_title("White Noise")
plt.plot(wn)
```

#### Correlation (Biased)

```python
Rub_ar=np.correlate(ar_data,ar_data, mode='full')
Rub_ar = Rub_ar / N
```

```python
fig, ax = plt.subplots()
ax.set_title("AR correlation")
indexes = np.arange(-N+1,N)
# ax.set_xlim(-10,10)
plt.stem(indexes,Rub_ar)
```

```python
Rub_wn=np.correlate(wn,wn, mode='full')
Rub_wn = Rub_wn / N
```

```python
fig, ax = plt.subplots()
ax.set_title("White Noise correlation")
indexes = np.arange(-N+1,N)
# ax.set_xlim(-10,10)
plt.stem(indexes,Rub_wn)
```
